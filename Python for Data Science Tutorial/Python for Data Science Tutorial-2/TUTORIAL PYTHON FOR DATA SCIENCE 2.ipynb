#### Question 2(g)
Consider following rules in classify the attendance:

    Rate > 90% : Good
    60% < Rate < 90% : Moderate
    Rate < 60% : Bad

Based on the grouping rule above, get the status for each of the staff and save the status in a new column `status` in the Data Frame


```python
# WRITE YOUR CODE HERE

def status(rate):   
    if rate > 90:
        status = 'Good'
    elif rate < 60:
        status = 'Bad'
    else:
        status = 'Moderate'
    return status

emp_attendance['status'] = emp_attendance['rate'].apply(status)
emp_attendance
```

*Breakdown of the code:*

1. The function `status` is defined with a single argument `rate`.

2. Inside the function, conditional statements are used to determine the attendance status based on the value of the `rate` argument.

3. If the `rate` is greater than 90, the status is set to 'Good'.

4. If the `rate` is less than 60, the status is set to 'Bad'.

5. If none of the above conditions are met, the status is set to 'Moderate'.

6. The function returns the calculated attendance status.

7. Outside the function, the `.apply()` method is used on the "rate" column of the `emp_attendance` DataFrame.

8. The `status` function is applied to each value in the "rate" column, resulting in the corresponding attendance status.

9. The resulting attendance statuses are assigned to a new column called "status" in the `emp_attendance` DataFrame.

*Therefore, when you run the code, it will add a new column named "status" to the `emp_attendance` DataFrame, containing the attendance status for each employee based on their attendance rate. The status is determined using the `status` function.*

#### Question 2(h)
Construct a function that returns the average attendance of all staff over the 14 days according to the department

```python
# WRITE YOUR CODE HERE

def AvgAttendDept(emp_attendance):
    return emp_attendance.groupby(['Department'],as_index=False).total.mean()

print(AvgAttendDept(emp_attendance))
```

*Breakdown of the code:*

1. The function `AvgAttendDept` is defined with the `emp_attendance` DataFrame as the input.

2. Inside the function, the `groupby()` method is used on the `emp_attendance` DataFrame to group the data by the "Department" column.

3. The `as_index=False` argument ensures that the "Department" column is not set as the index in the resulting grouped DataFrame.

4. The `.total.mean()` part calculates the mean value of the "total" column for each department.

5. The resulting grouped DataFrame with the average attendance per department is returned by the function.

6. Outside the function, the `AvgAttendDept` function is called with the `emp_attendance` DataFrame as the argument.

7. The returned grouped DataFrame is displayed using the `print()` function.
