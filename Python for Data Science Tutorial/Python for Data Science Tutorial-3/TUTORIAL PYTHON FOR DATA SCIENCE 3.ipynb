#### Question 3(i)
Create a function named eliminate_ref. This function will receive an array and perform these tasks:

#### Question 3(i(i))
Find whether there is any reference attached to any string in the array given. Commonly, any references can be detected by a square bracket containing a number like [5] as shown in figure below.

#### Question 3(i(ii))
If there is any reference attached, replace the word with the original word but without the reference. Set inplace=True when performing the replace method.
Test your function using these codes


The output should be as shown in figure below.


```python
# WRITE YOUR CODE HERE

import re

def eliminate_ref(arr):
    pattern = re.compile(r'\[\d+\]')
    for r in arr:
        matches = pattern.findall(r)
        if matches:
            for match in matches:
                replace_with = r.replace(match, "")
                print("Reference found: " + r + ", replacing it with: " + replace_with)
                arr.replace({r: replace_with}, inplace=True)

print("Processing Category")
eliminate_ref(tallest_3["Category"])

print("Processing Structure")
eliminate_ref(tallest_3["Structure"])

print("Processing Year built")
eliminate_ref(tallest_3["Year built"])

tallest_3
```

*Breakdown of the code:*

1. `import re` imports the regular expression module in Python.

2. `eliminate_ref(arr)` is a function that takes an array (`arr`) as input.

3. `pattern = re.compile(r'\[\d+\]')` compiles a regular expression pattern using the `re.compile()` function. The pattern is set to match square brackets containing one or more digits.

4. The function loops over each element `r` in the input array `arr`.

5. `matches = pattern.findall(r)` finds all matches of the regular expression pattern in the element `r` and stores them in the `matches` variable.

6. If there are matches (`if matches:`), the function enters a nested loop to iterate over each match.

7. `replace_with = r.replace(match, "")` replaces each match found in the element `r` with an empty string, effectively removing the reference.

8. The function prints a message indicating the reference found and the replacement made.

9. `arr.replace({r: replace_with}, inplace=True)` replaces the original element `r` with the modified `replace_with` value in the input array `arr`. The `inplace=True` argument ensures that the changes are made directly to the DataFrame.

10. The function is called multiple times, each time specifying a different column of the DataFrame `tallest_3` to process.

11. The modified DataFrame `tallest_3` is displayed.

*Therefore, when you run the code, it will apply the `eliminate_ref` function to the specified columns of the DataFrame `tallest_3` to remove references indicated by square brackets. References found in the columns "Category", "Structure", and "Year built" will be replaced with the modified values. The resulting DataFrame `tallest_3` will be displayed, showing the columns with the eliminated references.*
