#### Task 1(d)

The products in the sales data can be further grouped as follow:

a. Category A: facecream, facewash, moisturizer
b. Category B: toothpaste
c. Category C: bathingsoap, shampoo

Group the data accordingly and obtain the following information, then plot into two subplots:

1. total unit sold of each category for every month
2. percentage of total unit sold of each category for every month over the grand total of unit sold of that category.

```python
# WRITE YOUR CODE HERE

# Create a new column for the total units sold of each category
sales['Category A'] = sales[['facecream', 'facewash', 'moisturizer']].sum(axis=1)
sales['Category B'] = sales['toothpaste']
sales['Category C'] = sales[['bathingsoap', 'shampoo']].sum(axis=1)

# Calculate the total units sold for each category for every month
category_totals = sales.groupby('month_number')[['Category A', 'Category B', 'Category C']].sum()

# Calculate the percentage of total units sold for each category for every month over the grand total
category_totals['Category A %'] = (category_totals['Category A'] / sales['Category A'].sum()) * 100
category_totals['Category B %'] = (category_totals['Category B'] / sales['Category B'].sum()) * 100
category_totals['Category C %'] = (category_totals['Category C'] / sales['Category C'].sum()) * 100

# Plot the total units sold for each category for every month
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
category_totals[['Category A', 'Category B', 'Category C']].plot(kind='bar', ax=ax[0])
ax[0].set_title('Total Units Sold by Category')

# Plot the percentage of total units sold for each category for every month over the grand total
category_totals[['Category A %', 'Category B %', 'Category C %']].plot(kind='bar', ax=ax[1])
ax[1].set_title('Percentage of Total Units Sold by Category')

plt.show()
```

*Breakdown of the code:*

1. The code adds new columns to the `sales` DataFrame to calculate and store the total units sold for each category (`Category A`, `Category B`, `Category C`) by summing the corresponding product columns.

2. The `category_totals` DataFrame is created by grouping the `sales` DataFrame by the 'month_number' column and calculating the sum of units sold for each category.

3. The code calculates the percentage of total units sold for each category by dividing the units sold for each category by the sum of units sold for that category across all months.

4. Two subplots (`ax[0]` and `ax[1]`) are created using `plt.subplots()` to plot the total units sold and the percentage of total units sold for each category, respectively.

5. The total units sold for each category for every month are plotted as a bar chart in the first subplot (`ax[0]`) using the `plot()` function.

6. The percentage of total units sold for each category for every month over the grand total is plotted as a bar chart in the second subplot (`ax[1]`) using the `plot()` function.

7. Various `plt` functions are called to set titles for each subplot.

8. `plt.show()` displays the plot.

*Therefore, when you run the code, it will calculate the total units sold and the percentage of total units sold for different product categories over the months. It will then create two bar plots: one showing the total units sold by category, and the other showing the percentage of total units sold by category. These plots provide visual representations of the sales distribution and the relative contribution of each category to the overall sales.*

#### Task 2(b) 

From the above datasets, produce a chart (any appropriate chart) for each below:

a. products sold by each store
b. quantity sold against each product

```python
# Merge the sales and products dataframes 
merged_df = pd.merge(sales, products)

# Merge the merged_df and customers dataframes on the c_id column
merged_df = pd.merge(merged_df, customers, on='c_id')
merged_df
```

*Breakdown of the code:*

1. `merged_df = pd.merge(sales, products)` merges the `sales` and `products` DataFrames based on the common column(s) between them.

2. The resulting merged DataFrame is stored in `merged_df`.

3. `merged_df = pd.merge(merged_df, customers, on='c_id')` further merges the `merged_df` DataFrame with the `customers` DataFrame, using the 'c_id' column as the common column between them.

4. The final merged DataFrame is stored in `merged_df`.

*Therefore, when you run the code, it will merge the `sales` and `products` DataFrames based on their shared column(s), and then merge the resulting DataFrame with the `customers` DataFrame based on the 'c_id' column. The `merged_df` DataFrame will contain the combined data from all three DataFrames, allowing for analysis and further operations on the merged data.*

```python
#a. products sold by each store

store_products = merged_df.groupby(['store', 'product']).sum()['qty'].reset_index(name='qty')
store_products.pivot(index= 'store', columns= 'product', values='qty').plot(kind='bar', stacked=True)
plt.title('Products Sold by Each Store')
plt.xlabel('Store')
plt.xticks(rotation=0)
plt.ylabel('Quantity Sold')
plt.legend(title='Products')
plt.show()
```

*The code performs the following operations and creates a stacked bar plot:*

1. `store_products = merged_df.groupby(['store', 'product']).sum()['qty'].reset_index(name='qty')` groups the `merged_df` DataFrame by 'store' and 'product', calculates the sum of the 'qty' column, and creates a new DataFrame `store_products` with columns 'store', 'product', and 'qty'.

2. `store_products.pivot(index='store', columns='product', values='qty')` restructures the `store_products` DataFrame by pivoting it, using 'store' as the index, 'product' as the columns, and 'qty' as the values. This creates a new DataFrame with stores as rows, products as columns, and the corresponding quantity sold as values.

3. `.plot(kind='bar', stacked=True)` creates a stacked bar plot using the pivoted DataFrame. Each bar represents a store, and the height of each segment within the bar represents the quantity sold for each product category.

4. Various `plt` functions are called to set the title, x-axis label, x-tick labels, y-axis label, and legend for the plot.

5. `plt.show()` displays the plot.

*Therefore, when you run the code, it will calculate the total quantity sold for each product category in each store, create a stacked bar plot to visualize the data, and display the plot. The plot represents the products sold by each store, with the stacked bars showing the quantity sold for each product category. This visualization allows for comparisons between stores and product categories in terms of sales quantity.*

```python
#b.quantity sold against each product 

merged_df.groupby('product').sum()['qty'].plot.bar()

plt.title("Quantity sold against each product")
plt.xlabel("Products")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

*The code performs the following operations and creates a bar plot:*

1. `merged_df.groupby('product').sum()['qty']` groups the `merged_df` DataFrame by 'product', calculates the sum of the 'qty' column for each product, and returns a Series with the total quantity sold for each product.

2. `.plot.bar()` creates a bar plot using the `plot()` function of the resulting Series. Each bar represents a product, and the height of each bar corresponds to the total quantity sold for that product.


*Therefore, when you run the code, it will calculate the total quantity sold for each product in the `merged_df` DataFrame and create a bar plot to visualize the data. The plot represents the quantity sold against each product, allowing for comparisons and insights into the sales performance of different products.*

```python
#c.quantity and total sales against each product


merged_df['sales'] = merged_df.price*merged_df.qty

fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.4

product_qty = merged_df.groupby(['product']).agg({'qty': 'sum'})
product_qty.plot(kind='bar', color='red', ax=ax, width=width, position=1)

product_sales = merged_df.groupby(['product']).agg({'sales': 'sum'})
product_sales.plot(kind='bar', color='blue', ax=ax2, width=width, position=0)


ax.set_ylabel('Quantity Sold')
ax2.set_ylabel('Total Sales')
ax.set_xlabel('Products')
ax.set_xticklabels(["CPU", "Monitor", "RAM"],rotation =0)

ax.legend(["Quantity"],loc='upper left')
ax2.legend(["Total sales"],loc='upper right')

plt.xlim(-1,3)
plt.show()
```

*The code performs the following operations and creates a dual-axis bar plot:*

1. `merged_df['sales'] = merged_df.price*merged_df.qty` calculates the sales for each entry in the `merged_df` DataFrame by multiplying the 'price' and 'qty' columns, and assigns the result to a new 'sales' column.

2. `fig = plt.figure()` creates a new matplotlib figure.

3. `ax = fig.add_subplot(111)` creates a matplotlib axes object.

4. `ax2 = ax.twinx()` creates another axes object that shares the same x-axis as `ax`, allowing for dual-axis plotting.

5. `width = 0.4` sets the width of the bars in the plot.

6. `product_qty = merged_df.groupby(['product']).agg({'qty': 'sum'})` groups the `merged_df` DataFrame by 'product' and calculates the sum of the 'qty' column for each product, storing the result in the `product_qty` DataFrame.

7. `product_qty.plot(kind='bar', color='red', ax=ax, width=width, position=1)` creates a bar plot using the `plot()` function of the `product_qty` DataFrame. The bars are colored red and are plotted on `ax` with a specified width and position.

8. `product_sales = merged_df.groupby(['product']).agg({'sales': 'sum'})` groups the `merged_df` DataFrame by 'product' and calculates the sum of the 'sales' column for each product, storing the result in the `product_sales` DataFrame.

9. `product_sales.plot(kind='bar', color='blue', ax=ax2, width=width, position=0)` creates a bar plot using the `plot()` function of the `product_sales` DataFrame. The bars are colored blue and are plotted on `ax2` with a specified width and position.

10. Various `plt` functions are called to set the y-axis labels, x-axis label, x-tick labels, and legends for each axis.

11. `plt.xlim(-1,3)` sets the x-axis limits.

12. `plt.show()` displays the plot.

*Therefore, when you run the code, it will calculate the total quantity sold and the total sales for each product in the `merged_df` DataFrame. It will then create a dual-axis bar plot to visualize both the quantity sold and the total sales for each product. The red bars represent the quantity sold, and the blue bars represent the total sales. The plot allows for a comparison between the quantity and sales figures, providing insights into the relationship between them.*
